```python
import pandas as pd
```

```python
superstore = pd.read_csv("../data/raw/superstore.csv")

superstore.head(5)
```

```python
# Reviso el tipo de datos y cantidad de valores.
superstore.info()
```

```python
# Estadísticas descriptivas de los valores numéricos.
superstore.describe()
```

```python
# Calculaar la cantidad de valores nulos por columna.
superstore.isnull().sum()
```

```python
# Convertir las columnas de fecha de string a tipo datetime
superstore["Order Date"] = pd.to_datetime(superstore["Order Date"], dayfirst=True)
superstore["Ship Date"] = pd.to_datetime(superstore["Ship Date"], dayfirst=True)
```

```python
# Verifico a que ciudades y estados pertenecen los códigos postales nulos
superstore[superstore["Postal Code"].isnull()][["City", "State"]]
```

```python
# Reviso todos los datos cuya ciudad sea Burlington y el estado Vermont
superstore[
    (superstore["City"] == "Burlington") &
    (superstore["State"] == "Vermont")
]
```

```python
superstore["lead_time_days"] = (superstore["Ship Date"] - superstore["Order Date"]).dt.days
```

```python
superstore.head()
```

```python
superstore["lead_time_days"].describe()
```

```python
superstore[superstore["lead_time_days"] == 0][["Order Date", "Ship Date", "Ship Mode"]].head()
```

```python
superstore.groupby("Ship Mode")["lead_time_days"].agg(["mean", "median", "count"])
```

```python
superstore.groupby("Ship Mode")["Sales"].mean().sort_values(ascending=False)
```

```python
region_sales = (superstore.groupby("Region").agg(total_sales=("Sales", "sum"), avg_sales=("Sales", "mean"), 
                                                 orders=("Order ID", "nunique")).sort_values("total_sales", ascending=False))

region_sales
```

```python
city_sales = (superstore.groupby(["Region", "City"]).agg(total_sales=("Sales", "sum"), orders=("Order ID", "nunique")).sort_values("total_sales", ascending=False))

city_sales.head(10)
```

```python
customer_ltv = (superstore.groupby(["Customer ID", "Customer Name"]).agg(total_sales=("Sales", "sum"), orders=("Order ID", "nunique")).sort_values("total_sales", ascending=False))

customer_ltv.head(10)
```

```python
customer_ltv["avg_order_value"] = (customer_ltv["total_sales"] / customer_ltv["orders"])

customer_ltv.head(10)
```

```python
order_products = (superstore.groupby("Order ID").agg(num_products=("Product ID", "nunique"), total_sales=("Sales", "sum")))

order_products.head()
```

```python
order_products["order_type"] = (
    order_products["num_products"]
    .apply(lambda x: "Multi-product" if x > 1 else "Single-product")
)
```

```python
order_products["order_type"].value_counts()
```

```python
order_products["order_type"].value_counts(normalize=True) * 100
```

```python
order_products.groupby("order_type").agg(avg_sales=("total_sales", "mean"), median_sales=("total_sales", "median"),
    orders=("total_sales", "count"))
```

```python
superstore["order_month"] = superstore["Order Date"].dt.to_period("M")
```

```python
monthly_sales = (superstore.groupby("order_month")["Sales"].sum().reset_index().sort_values("order_month"))

monthly_sales.head()
```

```python
monthly_orders = (superstore.groupby("order_month")["Order ID"].nunique().reset_index(name="orders").sort_values("order_month"))

monthly_orders.head()
```

```python
monthly_aov = (superstore.groupby("order_month")["Sales"].mean().reset_index(name="avg_order_value").sort_values("order_month"))

monthly_aov.head()
```

```python
superstore["year"] = superstore["Order Date"].dt.year
superstore["month"] = superstore["Order Date"].dt.month

seasonality = (
    superstore
    .groupby("month")["Sales"]
    .mean()
    .reset_index()
)
```

```python
superstore[superstore["Order Date"].dt.month == 3].sort_values("Sales", ascending=False).head()
```

```python
top_products_volume = (superstore.groupby("Product Name")
    .agg(orders=("Order ID", "nunique"), total_sales=("Sales", "sum")).sort_values("orders", ascending=False)
)

top_products_volume.head(10)
```

```python
top_products_revenue = (superstore.groupby("Product Name")["Sales"]
    .sum().sort_values(ascending=False).head(10)
)

top_products_revenue
```

```python
superstore_processed = superstore.drop(columns=["Postal Code"])
```
